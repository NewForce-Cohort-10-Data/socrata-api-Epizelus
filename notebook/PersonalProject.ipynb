2. Now, use the API to get all available data for _Life expectancy at birth, total (years)_. This indicator has code "SP.DYN.LE00.IN". Again, convert the results to a DataFrame.

```python
endpoint2 = 'https://api.worldbank.org/v2/country/all/indicator/SP.DYN.LE00.IN?date=2017&format=json'

response2 = requests.get(endpoint2)
```

```python
response2
```

```python
res2 = response2.json()
```

```python
res2[1]
```

```python
lifeexp_df = pd.json_normalize(response2.json()[1])
lifeexp_df.columns
```

```python
lifeexp_df = lifeexp_df.drop(columns = ['unit', 'obs_status', 'decimal','indicator.id'])
```

```python
lifeexp_df = lifeexp_df.rename(columns = {'value' : 'Life Expectancy (Total Years)'})
```

```python
lifeexp_df.head()
```

3. Merge the two results DataFrames together. You may want to rename or drop columns prior to merging.

```python
gdpFrame['country.value'].unique()
```

```python
lifeexp_df['country.value'].unique()
```

```python
gdp_subset = gdpFrame[['countryiso3code', 'date', 'GDP Value', 'country.id', 'country.value']]
lifeexp_subset = lifeexp_df[['countryiso3code', 'date', 'Life Expectancy (Total Years)', 'country.id','country.value']]
```

```python
merged_df = pd.merge(
    gdp_subset,
    lifeexp_subset,
    on=['countryiso3code', 'date','country.id', 'country.value'], # Columns to match on
    how='outer'                     
)
```
